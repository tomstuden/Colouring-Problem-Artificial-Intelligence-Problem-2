```python
##Question One
##Resolving conflicts in 1 iteration

import networkx as nx
import random
import matplotlib.pyplot as plt

#Define Graph
graph = nx.Graph()
graph.add_edges_from([(1, 2), (1, 3), (1, 4), (1, 5), (2, 3), (2, 4), (2, 5), 
                      (3, 4), (3, 5), (4, 5), (5, 6), (5, 7), (6, 7), (6, 8), 
                      (7, 8), (7, 9), (8, 9), (8, 10), (9, 10), (9, 11), (10, 11), 
                      (10, 12), (11, 12), (11, 13), (12, 13), (12, 14), (13, 14), 
                      (13, 15), (14, 15), (14, 16), (15, 16), (15, 17), (16, 17), 
                      (16, 18), (17, 18), (17, 19), (18, 19), (18, 20), (19, 20), 
                      (19, 21), (20, 21), (20, 22), (21, 22),
                      (5, 15), (6, 14), (7, 13), (8, 12), (9, 11), (10, 16), (11, 17),
                      (12, 18), (13, 19), (14, 20), (15, 21), (16, 22), (2,21)])

# Assign initial colors and max colors number
numColors = 5
colors = {node: random.randint(1, numColors) for node in graph.nodes()}

# Function to update colors
def updateColors(node, colors):
    neighborColors = {colors[neighbor] for neighbor in graph.neighbors(node)}
    availableColors = set(range(1, numColors + 1)) - neighborColors
    colors[node] = random.choice(list(availableColors)) if availableColors else colors[node]

# Monitor conflicts and visualize graph evolution
conflictHistory = []
conflicted_nodes = {node for node in graph.nodes() if any(colors[node] == colors[neighbor] for neighbor in graph.neighbors(node))}
for iteration in range(1000):
    numConflictsResolved = 0
    for node_to_update in conflicted_nodes:
        neighborColors = {colors[neighbor] for neighbor in graph.neighbors(node_to_update)}
        if colors[node_to_update] in neighborColors:
            updateColors(node_to_update, colors)
            numConflictsResolved += 1
    conflictHistory.append(numConflictsResolved)
    if numConflictsResolved == 0:
        break  # Break the loop if there are no conflicts in the current iteration

# Visualize the graph
plt.figure(figsize=(12, 10))
nx.draw(graph, with_labels=True, node_color=[colors[node] for node in graph.nodes()])
plt.show()

# Plot conflicts over time
plt.plot(range(len(conflictHistory)), conflictHistory, marker='o')
plt.xlabel("Iteration")
plt.ylabel("Number of Conflicts Resolved")
plt.title("Number of Conflicts Resolved Over Single Iteration - Efficient Algorithm")
plt.grid(True)
plt.show()

# Count the number of nodes for each color
color_counts = {color: sum(1 for node_color in colors.values() if node_color == color) for color in range(1, numColors + 1)}

# Plot the bar chart
plt.figure(figsize=(8, 6))
plt.bar(color_counts.keys(), color_counts.values(), color=[f'C{color}' for color in color_counts.keys()])
plt.xlabel("Color")
plt.ylabel("Number of Nodes")
plt.title("Number of Nodes for Each Color - Solve")
plt.xticks(range(1, numColors + 1))
plt.grid(axis='y')
plt.show()

```

```python
import networkx as nx
import random
import matplotlib.pyplot as plt
import numpy as np

# Function to clear colors before coloring
def clear_colors(graph):
    for node in graph.nodes():
        if 'color' in graph.nodes[node]:
            del graph.nodes[node]['color']

#apply modified greedy coloring algorithm
def greedy_coloring(graph):
    clear_colors(graph)
    colors = {}  
    collisions = 0  
    
    for node in list(graph.nodes()):
        neighbor_colors = set(colors[neighbor] for neighbor in graph.neighbors(node) if neighbor in colors)
        
        color = 1
        while color in neighbor_colors:
            collisions += 1
            color += 1
        
        colors[node] = color
    
    return colors, len(set(colors.values())), collisions

#apply modified greedy coloring traversing nodes in reverse
def final_node_greedy_coloring(graph):
    clear_colors(graph)
    colors = {}  
    collisions = 0  
    
    for node in graph.nodes():
        colors[node] = 1
    
    nodes = list(reversed(list(graph.nodes())))
    
    for node in nodes:
        neighbor_colors = {colors[neighbor] for neighbor in graph.neighbors(node) if neighbor in colors}
        
       
        color = 2  
        while color in neighbor_colors:
            color += 1
        
        # Assign the color to the node
        colors[node] = color
        if color != 2: 
            collisions += 1  
    
    return colors, len(set(colors.values())), collisions

#coloring algorithm shuffling nodes randomly
def random_coloring(graph):
    clear_colors(graph)
    colors = {} 
    nodes = list(graph.nodes())
    random.shuffle(nodes) 
    collisions = 0  
    
    for node in nodes:
        neighbor_colors = set(colors[neighbor] for neighbor in graph.neighbors(node) if neighbor in colors)
        
        if neighbor_colors:
            available_colors = set(range(1, len(graph) + 1)) - neighbor_colors
            color = random.choice(list(available_colors))
            collisions += 1
        else:
            color = 1
        
        colors[node] = color
    
    return colors, len(set(colors.values())), collisions



#create graphs using num nodes list
def create_graph(num_nodes):
    m = 2  # Parameter for Barabasi-Albert model
    scale_free_graph = nx.barabasi_albert_graph(num_nodes, m)
    num_edges = len(scale_free_graph.edges)
    p = num_edges / (num_nodes * (num_nodes - 1) / 2) 
    random_graph = nx.fast_gnp_random_graph(num_nodes, p)
    return scale_free_graph, random_graph

#plotting colors used
def bar_chart_colors_used(graphs, coloring_functions, titles, graph_types, node_sizes):
    fig, axes = plt.subplots(nrows=len(graph_types), figsize=(10, 8))

    for i, graph_type in enumerate(graph_types):
        colors_used = [[] for _ in range(len(coloring_functions))]
        for j, graph_list in enumerate(graphs[i]):
            for k, coloring_func in enumerate(coloring_functions):
                _, num_colors, _ = coloring_func(graph_list)
                colors_used[k].append(num_colors)
        
        num_bars = len(coloring_functions)
        num_nodes = len(node_sizes)
        bar_width = 0.35 / num_bars 
        
        for k, coloring_func in enumerate(coloring_functions):
            x_pos = np.arange(num_nodes) + k * bar_width 
            axes[i].bar(x_pos, colors_used[k], label=coloring_func.__name__, width=bar_width, align='center', alpha=0.7, edgecolor='black', linewidth=1, color=plt.cm.tab10(k))
        
        axes[i].set_title(graph_type)
        axes[i].set_xticks(np.arange(num_nodes) + bar_width * (num_bars - 1) / 2)
        axes[i].set_xticklabels([f"{len(graph_list.nodes)} Nodes" for graph_list in graphs[i]])
        axes[i].set_xlabel("Graph Size")
        axes[i].set_ylabel("Number of Colors")
        axes[i].legend()

    plt.tight_layout()
    plt.show()
    
#line plot of collisions detected over each graph, coloring function, number of nodes
def line_chart_collisions(graphs, coloring_functions, titles, graph_types):
    num_collisions = {graph_type: {func.__name__: [] for func in coloring_functions} for graph_type in graph_types}
    
    for graph_type, graph_list in zip(graph_types, graphs):
        for graph in graph_list:
            for coloring_func in coloring_functions:
                _, _, collisions = coloring_func(graph)
                num_collisions[graph_type][coloring_func.__name__].append(collisions)
    
    plt.figure(figsize=(10, 6))
    for graph_type, data in num_collisions.items():
        for func_name, collisions in data.items():
            plt.plot(collisions, label=f"{graph_type} - {func_name}")
    
    plt.xticks(range(len(graph_list)), [f"{len(graph.nodes)}" for graph in graph_list])
    plt.xlabel("Graph Size")
    plt.ylabel("Number of Collisions")
    plt.title("Number of Collisions Detected by Coloring Algorithm")
    plt.legend()
    plt.grid(True)
    plt.show()

#colors used line plot across all graphs and coloring functions
def line_chart_colors_used(graphs, coloring_functions, titles, graph_types):
    num_colors_used = {graph_type: {func.__name__: [] for func in coloring_functions} for graph_type in graph_types}
    
    for graph_type, graph_list in zip(graph_types, graphs):
        for graph in graph_list:
            for coloring_func in coloring_functions:
                colors, num_colors, _ = coloring_func(graph)
                num_colors_used[graph_type][coloring_func.__name__].append(num_colors)
    
    plt.figure(figsize=(10, 6))
    for graph_type, data in num_colors_used.items():
        for func_name, num_colors in data.items():
            plt.plot(num_colors, label=f"{graph_type} - {func_name}")
    
    plt.xticks(range(len(graph_list)), [f"{len(graph.nodes)}" for graph in graph_list])
    plt.xlabel("Graph Size")
    plt.ylabel("Number of Colors Used")
    plt.title("Number of Colors Used by Coloring Algorithm")
    plt.legend()
    plt.grid(True)
    plt.show()

##running
#define graph types
graph_types = ["Barabasi", "Random"]

#set node size list
node_sizes = [25, 50, 75, 100, 150, 200, 300]

#list of coloring functions
coloring_functions = [greedy_coloring, final_node_greedy_coloring, random_coloring]

#creating graphs
barabasi_graphs = [create_graph(size)[0] for size in node_sizes]
random_graphs = [create_graph(size)[1] for size in node_sizes]

# Titles for the plots
titles = [func.__name__.replace("_", " ").title() for func in coloring_functions]

# Plot bar charts of colors used by each function on each graph
bar_chart_colors_used([barabasi_graphs, random_graphs], coloring_functions, titles, graph_types, node_sizes)

# Plot line chart of number of collisions detected by each function across each graphs
line_chart_collisions([barabasi_graphs, random_graphs], coloring_functions, titles, graph_types)

# Plot line chart of number of colors used by each function across each graphs
line_chart_colors_used([barabasi_graphs, random_graphs], coloring_functions, titles, graph_types)

#create 100 node graph of both types
def create_100_node_graph():
    num_nodes = 100
    m = 2  
    scale_free_graph = nx.barabasi_albert_graph(num_nodes, m)
    num_edges = len(scale_free_graph.edges)
    p = num_edges / (num_nodes * (num_nodes - 1) / 2)  
    random_graph = nx.fast_gnp_random_graph(num_nodes, p)
    return scale_free_graph, random_graph

#plot colors used on 100 node graph
def plot_colors_used_per_function(graph, coloring_functions, title):
    fig, ax = plt.subplots(figsize=(8, 6))

    num_colors_used = []
    for coloring_func in coloring_functions:
        _, num_colors, _ = coloring_func(graph)
        num_colors_used.append(num_colors)

    x_pos = np.arange(len(coloring_functions)) 
    colors = ['blue', 'green', 'orange']
    ax.bar(x_pos, num_colors_used, width=0.5, align='center', alpha=0.7, edgecolor='black', linewidth=1, color=colors)
        
    ax.set_xticks(x_pos)
    ax.set_xticklabels([func.__name__ for func in coloring_functions], rotation=45, ha='right')
    ax.set_xlabel("Coloring Function")
    ax.set_ylabel("Number of Colors")
    ax.set_title(f"Number of Colors Used by Each Coloring Function on - {title} : {graph}")

    plt.tight_layout()
    plt.show()

barabasi_100_graph, random_100_graph = create_100_node_graph()

plot_colors_used_per_function(barabasi_100_graph, coloring_functions, "Barabasi")

plot_colors_used_per_function(random_100_graph, coloring_functions, "Random")

# Function to plot degree distribution
def plot_degree_distribution(graphs, graph_types, titles):
    max_degree = 0
    for graph_list in graphs:
        degrees = [list(dict(graph.degree()).values()) for graph in graph_list]
        max_degree = max(max_degree, max([max(degree) for degree in degrees]))

    fig, axes = plt.subplots(1, len(graphs), figsize=(15, 5))

    for i, graph_list in enumerate(graphs):
        degrees = [list(dict(graph.degree()).values()) for graph in graph_list]
        axes[i].hist(degrees, bins=np.arange(max_degree + 2) - 0.5, density=True, alpha=0.7, label=graph_types[i])
        axes[i].set_title(titles[i])
        axes[i].set_xlabel("Degree")
        axes[i].set_ylabel("Density")
        axes[i].legend()

    plt.tight_layout()
    plt.show()

# Plot degree distribution of Barabasi and Random graphs
plot_degree_distribution([barabasi_graphs, random_graphs], graph_types, ["Barabasi Graphs", "Random Graphs"])


```
